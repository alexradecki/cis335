# Cleaning and Model Prep

```python
train_df = train_df.drop(['id'] , axis = 1)
train_df = train_df.drop(['Unnamed: 32'] , axis = 1)

test_df = test_df.drop(['id'] , axis = 1)
test_df = test_df.drop(['Unnamed: 32'] , axis = 1)
```

Converts diagnosis into a numeric binary value

```python
train_df.loc[train_df["diagnosis"] == "M", "diagnosis"] = 1
train_df.loc[train_df["diagnosis"] == "B", "diagnosis"] = 0

test_df.loc[test_df["diagnosis"] == "M", "diagnosis"] = 1
test_df.loc[test_df["diagnosis"] == "B", "diagnosis"] = 0
```

Puts all predictor variables into list called "features"

```python
features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean','smoothness_mean',
            'compactness_mean', 'concavity_mean','concave points_mean', 'symmetry_mean',
            'fractal_dimension_mean','radius_se', 'texture_se', 'perimeter_se', 'area_se',
            'smoothness_se','compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
            'fractal_dimension_se', 'radius_worst', 'texture_worst','perimeter_worst', 'area_worst',
            'smoothness_worst','compactness_worst', 'concavity_worst', 'concave points_worst','symmetry_worst',
            'fractal_dimension_worst']
```

Separating datasets into diagnosis and feature values and verifies diagnosis values are integers

```python
y_train = train_df['diagnosis']
y_test = test_df['diagnosis']

x_train = train_df[features]
x_test = test_df[features]

y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

Performs z-score and min-max normalization on the feature values

```python
x_train_zscore = (x_train - x_train.mean()) / (x_train.std())
x_test_zscore = (x_test - x_train.mean()) / (x_train.std())

x_train_minmax = (x_train - x_train.min()) / (x_train.max() - x_train.min())
x_test_minmax = (x_test - x_train.min()) / (x_train.max() - x_train.min())
```

# Model Fitting and Training

Imports decision tree classifier from the sklearn package and creates 18 instances of it; One for every variation of decision tree depth, splitter, and normalization process.


```python
from sklearn.tree import DecisionTreeClassifier
decision_tree_5_best_1 = DecisionTreeClassifier(max_depth=5, splitter="best")
decision_tree_5_best_2 = DecisionTreeClassifier(max_depth=5, splitter="best")
decision_tree_5_best_3 = DecisionTreeClassifier(max_depth=5, splitter="best")

decision_tree_5_random_1 = DecisionTreeClassifier(max_depth=5, splitter="random")
decision_tree_5_random_2 = DecisionTreeClassifier(max_depth=5, splitter="random")
decision_tree_5_random_3 = DecisionTreeClassifier(max_depth=5, splitter="random")

decision_tree_6_best_1 = DecisionTreeClassifier(max_depth=6, splitter="best")
decision_tree_6_best_2 = DecisionTreeClassifier(max_depth=6, splitter="best")
decision_tree_6_best_3 = DecisionTreeClassifier(max_depth=6, splitter="best")

decision_tree_6_random_1 = DecisionTreeClassifier(max_depth=6, splitter="random")
decision_tree_6_random_2 = DecisionTreeClassifier(max_depth=6, splitter="random")
decision_tree_6_random_3 = DecisionTreeClassifier(max_depth=6, splitter="random")

decision_tree_7_best_1 = DecisionTreeClassifier(max_depth=7, splitter="best")
decision_tree_7_best_2 = DecisionTreeClassifier(max_depth=7, splitter="best")
decision_tree_7_best_3 = DecisionTreeClassifier(max_depth=7, splitter="best")

decision_tree_7_random_1 = DecisionTreeClassifier(max_depth=7, splitter="random")
decision_tree_7_random_2 = DecisionTreeClassifier(max_depth=7, splitter="random")
decision_tree_7_random_3 = DecisionTreeClassifier(max_depth=7, splitter="random")
```

Fits models by training on their respective datasets on our target variable diagnosis

```python
decision_tree_5_best_1.fit(x_train, y_train)
decision_tree_5_best_2.fit(x_train_minmax, y_train)
decision_tree_5_best_3.fit(x_train_zscore, y_train)

decision_tree_5_random_1.fit(x_train, y_train)
decision_tree_5_random_2.fit(x_train_minmax, y_train)
decision_tree_5_random_3.fit(x_train_zscore, y_train)

decision_tree_6_best_1.fit(x_train, y_train)
decision_tree_6_best_2.fit(x_train_minmax, y_train)
decision_tree_6_best_3.fit(x_train_zscore, y_train)

decision_tree_6_random_1.fit(x_train, y_train)
decision_tree_6_random_2.fit(x_train_minmax, y_train)
decision_tree_6_random_3.fit(x_train_zscore, y_train)

decision_tree_7_best_1.fit(x_train, y_train)
decision_tree_7_best_2.fit(x_train_minmax, y_train)
decision_tree_7_best_3.fit(x_train_zscore, y_train)

decision_tree_7_random_1.fit(x_train, y_train)
decision_tree_7_random_2.fit(x_train_minmax, y_train)
decision_tree_7_random_3.fit(x_train_zscore, y_train)
```

For each model, create a dataset that holds the values of the model's predictions when using its respective test dataset

```python
predictions_5_best_1 = decision_tree_5_best_1.predict(x_test)
predictions_5_best_2 = decision_tree_5_best_2.predict(x_test_minmax)
predictions_5_best_3 = decision_tree_5_best_3.predict(x_test_zscore)

predictions_5_random_1 = decision_tree_5_random_1.predict(x_test)
predictions_5_random_2 = decision_tree_5_random_2.predict(x_test_minmax)
predictions_5_random_3 = decision_tree_5_random_3.predict(x_test_zscore)

predictions_6_best_1 = decision_tree_6_best_1.predict(x_test)
predictions_6_best_2 = decision_tree_6_best_2.predict(x_test_minmax)
predictions_6_best_3 = decision_tree_6_best_3.predict(x_test_zscore)

predictions_6_random_1 = decision_tree_6_random_1.predict(x_test)
predictions_6_random_2 = decision_tree_6_random_2.predict(x_test_minmax)
predictions_6_random_3 = decision_tree_6_random_3.predict(x_test_zscore)

predictions_7_best_1 = decision_tree_7_best_1.predict(x_test)
predictions_7_best_2 = decision_tree_7_best_2.predict(x_test_minmax)
predictions_7_best_3 = decision_tree_7_best_3.predict(x_test_zscore)

predictions_7_random_1 = decision_tree_7_random_1.predict(x_test)
predictions_7_random_2 = decision_tree_7_random_2.predict(x_test_minmax)
predictions_7_random_3 = decision_tree_7_random_3.predict(x_test_zscore)
```
